```python
import numpy as np
x_train=np.load('../../../datasets/EYALEB/d_eyaleb.npy')
x_train=x_train.reshape(x_train.shape[0]*x_train.shape[1],784)
y=np.load('../../../datasets/EYALEB/eyalebpredictions.npy')
```

```python
from sklearn.manifold import TSNE
%matplotlib inline
import matplotlib.pyplot as plt
%pylab inline
import seaborn as sns

figure(figsize=(10, 6))
subplot(121)

X_tsne = TSNE(n_components=2,  verbose=1, random_state=123).fit_transform(x_train)

ax=sns.scatterplot(x=X_tsne[:,0], y=X_tsne[:,1], hue=y.tolist(), marker='.',s=8,
                palette=sns.color_palette("hls", 38),
                data=x_train,legend = False)
ax.axis('off')
#plt.savefig('SDSCeyalebc38.pdf') 
```

```python
from sklearn.manifold import TSNE
%matplotlib inline
import matplotlib.pyplot as plt
%pylab inline
import seaborn as sns

y_ids = range(len(y))
X_tsne = TSNE(n_components=2,  verbose=1, random_state=123).fit_transform(x_train)
figure(figsize=(6, 3))
subplot(121)
colors = 'lightgreen','g', 'b', 'k', 'pink', 'purple', 'gray', 'red','c','y'

for i, c, label in  zip(y_ids,colors,np.unique(y)):
    plt.scatter(X_tsne[y == i, 0], X_tsne[y == i, 1], label=label,c=c, marker='.',s=8)
    plt.axis('off')
#plt.savefig('SDSCeyalebc10.pdf') 
```

```python
x_train=np.load('../../../datasets/COIL100/d_coil100.npy')
x_train=x_train.reshape(x_train.shape[0]*x_train.shape[1],3000)
y=np.load('../../../datasets/COIL100/coil100predictions.npy')
```

```python
x=np.array(x_train)
y=np.array(y)
xtr=[]
ytr=[]
for i in range(len(y)):
    if y[i]==10:
        xtr.append(x[i])
        ytr.append(y[i])
    elif y[i]==1:
        xtr.append(x[i])
        ytr.append(y[i])
    elif y[i]==2:
        xtr.append(x[i])
        ytr.append(y[i])
    elif y[i]==3:
        xtr.append(x[i])
        ytr.append(y[i])
    elif y[i]==4:
        xtr.append(x[i])
        ytr.append(y[i])
    elif y[i]==5:
        xtr.append(x[i])
        ytr.append(y[i])
    elif y[i]==6:
        xtr.append(x[i])
        ytr.append(y[i])
    elif y[i]==7:
        xtr.append(x[i])
        ytr.append(y[i])
    elif y[i]==8:
        xtr.append(x[i])
        ytr.append(y[i])
    elif y[i]==9:
        xtr.append(x[i])
        ytr.append(y[i])
    else:
        continue
x_train=xtr
y=ytr

x=np.array(x)
y=np.array(y)
y=y-1
y=y.reshape(len(y),)

y_ids = range(len(y))
X_tsne = TSNE(n_components=2,  verbose=1, random_state=123).fit_transform(x_train)
figure(figsize=(6, 3))
subplot(121)
colors = 'lightgreen','g', 'b', 'k', 'pink', 'purple', 'gray', 'red','c','y'

for i, c, label in  zip(y_ids,colors,np.unique(y)):
    plt.scatter(X_tsne[y == i, 0], X_tsne[y == i, 1], label=label,c=c, marker='.',s=8)
    plt.axis('off')
#plt.savefig('SDSCcoil100c10.pdf') 
```
